```python
import numpy as np
import sympy
from sympy import Symbol
sympy.init_printing(use_latex='mathjax')
```

```python
x = sympy.symbols("x")
a = sympy.symbols("a")
sympy.integrate(sympy.exp(-a*x**2),(x,-sympy.oo,sympy.oo))
```

```python
x = sympy.symbols("x")
a = sympy.symbols("a")
sympy.integrate(x*sympy.exp(-a*x**2),(x,-sympy.oo,sympy.oo))
```

```python
x = sympy.symbols("x")
a = sympy.symbols("a")
sympy.integrate(x**2*sympy.exp(-a*x**2),(x,-sympy.oo,sympy.oo))
```

```python
x = sympy.symbols("x")
a = sympy.symbols("a")
sympy.integrate(x**4*sympy.exp(-a*x**2),(x,-sympy.oo,sympy.oo))
```

```python
from sympy import *
t = symbols('t')
x = Function('x')
x(t)
```
