```python
import random
import numpy as np
import matplotlib.pyplot as plt

def average_E(Beta,n):
    states = np.ones(n)
    for t in range(n*100):
        i = random.randint(0,n-1)
        dE = 0
        if random.random() < 1/2:
            dE = +1
        else:
            dE = -1
            
        if dE < 0:  
            states[i] = states[i] - 1 
        elif random.random() < np.exp(-Beta * dE):
            states[i] = states[i] + 1
        if states[i] == 0:
            states[i] = 1
    return states
```

```python
temp = np.arange(0,20,1)
y = [sum(average_E(1/Beta,1000) ** 2) for Beta in temp]
plt.plot(temp,y)
```

```python
temp = np.arange(1,10,1)
y = [sum(average_E(Beta,1000) ** 2) for Beta in temp]
plt.plot(temp,y)
```

```python
def Distribution(Beta,n):
    ct = average_E(Beta,n)
    Nb = np.zeros(10)
    for t in range(1,10):
        Nb[t] = np.count_nonzero(ct == t)
    return Nb
```

```python
plt.plot(Distribution(1,1000),label='Beta=1')
plt.plot(Distribution(2,1000),label='Beta=2')
plt.plot(Distribution(3,1000),label='Beta=3')
plt.plot(Distribution(4,1000),label='Beta=4')
plt.legend()

plt.xlabel("states")
plt.ylabel('Particles')
plt.show()
```
