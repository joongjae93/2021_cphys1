```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def Leibniz(a):
    f = 1
    for n in range(1,a):
        if n % 4 == 1:
            f += (1*(-1)**(n))/(n+2)
        elif n % 4 == 3:
            f += (1*(-1)**(n-1))/(n+2)
    return f
    
x = np.arange(0,100,2)
y = [Leibniz(s) for s in range(1,101) if s % 2 == 1]
plt.xlabel('n')
plt.plot(x,y,label='Leibniz formula for π')
plt.hlines(np.pi/4,0,100,color='orange',label='π/4')
plt.legend()
plt.show()
```

```python
def LogisticMap(f, r):
    x,y = [],[]
    for i in range(1,101):
        f = r*f*(1-f)
        x.append(i)
        y.append(f)


    plt.xlabel('n') 
    plt.ylabel('fn')
    plt.plot(x, y)
    plt.show()

LogisticMap(0.5, 0.5)
LogisticMap(0.5, 0.95)
LogisticMap(0.5, 2)
LogisticMap(0.5, 3.2)
LogisticMap(0.5, 3.8)
LogisticMap(0.501, 3.8)
```

```python
def sinc(n):
    x = np.linspace(-10,10,n)
    y = np.sin(x)/x
    
    plt.plot(x, y)

sinc(100)
```

```python
def sinc_Taylor(x,n):
    f = 0
    for i in range(n+1):
        f = f + ((-1)**i * x**(2*i))/(math.factorial(2*i+1))
    return f

    
x = np.arange(-10,10,0.1)
plt.plot(x,sinc_Taylor(x,1), color = 'purple')
plt.plot(x,sinc_Taylor(x,4), color = 'blue')
plt.plot(x,sinc_Taylor(x,9), color = 'red')
plt.plot(x,sinc_Taylor(x,16), color = 'green')
plt.axis([-10,10,-1,1])
plt.show()
```
