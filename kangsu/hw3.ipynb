# 1. Leibniz Pi formula

$$\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots 
\end{align}$$
이 식은 라이프니츠의 파이 공식이다.

다음의 식을 구현하는 함수를 만들어라.


```python
def Logistic(n):
    rfn=-1
    out=0
    for i in range(0, n):
        rfn=rfn+2
        fn=((-1)**(i))/rfn
        out=fn+out
    return out
             


```

$$\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}$$
이 식을 이용하여 아래처럼 그림을 그려라.

```python
import math    
fp=[]
n=100
u=math.pi

for j in range(1, n):
   
    fp.append(Logistic(j))
plt.plot(fp,'-',label='Leibniz formula for π')
plt.plot([0,100],[u/4,u/4],'-',label="π")
plt.xlabel('n')
plt.legend()

```

# 2. Logistic Map

로지스틱 맵은 다음처럼 주어진다.$$ f_{n+1} = r f_n (1- f_n). $$

$f_n$을 구하는 함수를 작성하여라.


```python
def Logistic(r,f,n1):
    f2=f
    for i in range(0, n1):
        f2=r*f2*(1-f2)
        
    return f2
```

작성한 함수를 이용하여 다음의 조건에서 $n=1$에서 $100$까지 로지스틱 맵의 값을 구하고 그래프로 그려라. 즉, x축은 $n$, y축은 $f_n$인 그래프를 그려라.

$r=0.5$, $f_0$=0.5

$r=0.95$, $f_0$=0.5

$r=2$, $f_0$=0.5

$r=3.2$, $f_0$=0.5

$r=3.8$, $f_0$=0.5

$r=3.8$, $f_0$=0.501

```python
r0=[0.5,0.95,2,3.2,3.8,3.8]
f0=[0.5,0.5,0.5,0.5,0.5,0.501]
fp=[0]*100
na=100

for k in range(0, 6):
    r = r0[k]
    f = f0[k]
    for j in range(1, na+1):
        fp[j-1]=(Logistic(r,f,j))
    plt.figure(figsize=(20, 20))
    plt.subplot(611+k)
    plt.plot(fp,'-')
    plt.ylabel('f(n)')
    plt.xlabel('n \n \n r=%.3f f=%.3f \n'%(r,f))
plt.show()
```

# 3. sinc function

싱크 함수를 그려보아라. 싱크함수는 다음처럼 주어진다.$$ sinc(x)  = \frac{\sin x}{x}. $$

```python
a=np.arange(-100,100,0.01)
b=[np.sin(x)/x for x in a]
plt.plot(a,b)
```

싱크 함수를 테일러 전개하면 다음과 같다.$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$위의 식의 우변을 $n$을 바꾸어 가면서 그래프로 그리고 싱크함수와 비교하여라.


```python

def Logistic(n):
    c=(((-1)**n)*a**(2*n))/math.factorial(2*n+1)
    return c

a=np.arange(1,40,0.01)

b=[np.sin(x)/x for x in a]



k=10

for i in range(0,k):
    f=a*0
    n=i*5
    for j in range(0,n):
        f=f+Logistic(j)
    plt.plot(a,f)
    plt.plot(a,b)
    plt.ylim(-0.25,0.2)
    plt.xlabel('n \n \n n=%d'%(n))
    plt.show()


    
```
