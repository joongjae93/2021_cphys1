#  1. Vibration in a one-dimensional system

질량이 $m$인 $N$개의 동일한 물체가 일렬로 동일한 용수철에 매달려 있다. $i$번째 물체의 운동방정식은 다음처럼 주어진다.

$$m \frac{d^2 x_i}{dt^2} = k(x_{i+1}-x_{i}) - k(x_i-x_{i-1})$$

한쪽 끝에서 연결된 용수철에 주기적인 힘($F$)을 가하고 있다. 그러면 양 끝단에 물체에 운동방정식은 다음과 같다.

$$m \frac{d^2 x_1}{dt^2} = k(x_{2}-x_{1}) + F \\
m \frac{d^2 x_N}{dt^2} = -k(x_{N}-x_{N-1})$$

만약 주기적인 힘이 $F=C e^{i wt}$로 주어지면 $x$는 다음의 $x_i(t)=a_i e^{iwt}$해를 가진다. 정리하면 다음과 같은 방정식을 얻을 수 있다.
$$(d -k)a_1 -k a_2 = C, \\
d a_i - k a_{i-1}- k a_{i+1} = 0, \\
(d - k)a_N - k x_{N-1} =0$$
여기에서 $d=2k-mw^2$이다.

$$Ma=b $$$$
M=
\begin{pmatrix}
d-k & -k & 0 & 0 & 0 & \cdots \\
-k & d & -k & 0 & 0 & \cdots \\
0 & -k & d & -k & 0 & \cdots \\
\cdots &\cdots &\cdots &\cdots &\cdots & \cdots  \\
0 & 0 & 0& 0& -k & d-k 
\end{pmatrix}, \quad
a=
\begin{pmatrix}
a_1 \\ 
a_2 \\ 
a_3 \\ 
\cdots \\ 
a_N  
\end{pmatrix}, \quad
b=
\begin{pmatrix}
C \\ 
0 \\ 
0 \\ 
\cdots \\ 
0 
\end{pmatrix}
$$

$m=1$, $k=1$, $C=1$, $N=40$으로 고정하세요.
$w$ (즉, d)를 바꾸어가면서 $a_i$를 구하세요.
이를 그래프로 표현하고 의미를 설명하세요. x축이 i (1~N), y축이 $a_i$
40X40 행렬을 직접 입력하는 것을 금지하지는 않지만 그렇게 하지 않기를 바랍니다.
np.zeros(n) 함수는 크기가 n이며 모든 성분이 0인 배열을 만드는 명령어입니다.

```python
#w=1인 경우
```

```python
m,k,w=1,1,1

def d(w):
    return (2*k)-(m*(w**2))


import numpy as np

ma1 = np.eye(40,40,0, dtype= float)
for i in range(0,40):
    for j in range(0,40):
        if i == 0 and j == 0:
            ma1[i][j] = d(w)-k
            
        elif i == 0 and j == 1:
            ma1[i][j] =-k
            
        elif i == 39 and j == 38:
            ma1[i][j] =-k
            
        elif i == 39 and j == 39:
            ma1[i][j] = d(w)-k
            
        elif i == j+1:
            ma1[i][j] =-k
            
        elif i == j:
            ma1[i][j] = d(w)
            
        elif j == i+1:
            ma1[i][j] =-k
            
for i in range(40):
    for j in range(40):
        if ma1[i][j] == '':
            ma1[i][j] = '0'
print(ma1)

M_inv=np.linalg.inv(ma1)
print(M_inv)
```

```python
C=1

import numpy as np
mb = np.eye(40,1,0, dtype= float)
for i in range(0,40):
    for j in range(1):
        if i == 0:
            mb[i][j] = C
        else:
            mb[i][j] = 0
print(mb)

k1=np.dot(M_inv,mb)
```
