```python
#Vibration in 1-D system
import numpy as np
import matplotlib.pyplot as plt
```

```python
#w=1일때 그래프
x = np.zeros((40,40))
m = 1
k = 1
w = 1
c = 1
d = 2*k-m*w**2

for i in range(1,40):
    x[i,i] = d
    x[i,i-1] = -k
    x[i-1,i] = -k

x[0,0]=d-k
x[39,39]=d-k
b = np.zeros(40)
b[0] = 1
x_a = np.linalg.solve(x,b)
xdata= np.arange(1,41,1)
plt.plot(xdata,x_a,'-o',label= 'w=1')
plt.xlabel('$i$')
plt.ylabel('$a_i$');
plt.legend()

#w=3일때 그래프

x = np.zeros((40,40))
m = 1
k = 1
w = 3
c = 1
d = 2*k-m*w**2

for i in range(1,40):
    x[i,i] = d
    x[i,i-1] = -k
    x[i-1,i] = -k

x[0,0]=d-k
x[39,39]=d-k
b = np.zeros(40)
b[0] = 1
x_a = np.linalg.solve(x,b)
xdata= np.arange(1,41,1)
plt.plot(xdata,x_a,'-o',label = 'w=3')
plt.legend();
```
