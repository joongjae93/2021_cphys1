```python
import numpy as np
import matplotlib.pyplot as plt
import random

#1.Travelling salesman
#x=[random.randint(0,10) for i in range(10)]
#y=[random.randint(0,10) for i in range(10)]
x=[0, 10, 9, 5, 10, 5, 4, 2, 8, 9]
y=[10, 0, 6, 8, 4, 3, 2, 7, 3, 2]
plt.plot(x,y,'o')
```

```python
plt.plot(x,y)
```

```python
def dist(x,y):
    z=0
    for i in range(9):
        z+=np.sqrt((x[i+1]-x[i])**2+(y[i+1]-y[i])**2)
    return z
dist(x,y)
```

```python
def de(i,j,x,y):
    nx, ny = x[:], y[:]
    nx[i], nx[j] = nx[j], nx[i]
    ny[i], ny[j] = ny[j], ny[i]
    return dist(nx,ny)-dist(x,y)
```

```python
import random

def swap(x,y,b):
    i=random.randint(0,9)
    j=random.randint(0,9)
    delta=de(i,j,x,y)
    if(delta<0):
        x[i], x[j] = x[j], x[i]
        y[i], y[j] = y[j], y[i]
    else:
        if(random.random()<np.exp(-b*delta)):
            x[i], x[j] = x[j], x[i]
            y[i], y[j] = y[j], y[i]
    return x,y
```

```python
Tmax=10
traj=[]
co = [0]

for i in range(7000):
    T=Tmax*np.exp(-i/10) # cooling  
    swap(x,y,1/T)
    d= dist(x,y)
    traj=np.append(traj,d)
    co=np.append(co,i+1)
```
