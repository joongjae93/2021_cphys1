```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
#3. Damped Pendulum
w = 1
b = 0.1
state0 = [1.0,0.0] # (x0 , v0)
t = np.arange(0.0,50.0,0.01)
def f(state,t):
    x,v = state
    return v, -2*b*v-w**2*x
states = odeint(f,state0,t)
plt.plot(states[:,0])
```

```python
#4. Forced Pendulum
#V 
#V = 5
state0 = [1,0]  # [theta0, omega0]
 = 0.1
gL = 1
A = 2
v = 5
t = np.arange(0,100,0.1)
def f(state, t):
    , w = state
    return w, -*w-gL*np.sin()+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 5')
plt.legend()

#V = 30
state0 = [1,0]  # [theta0, omega0]
 = 0.1
gL = 1
A = 2
v = 30
t = np.arange(0,100,0.1)
def f(state, t):
    , w = state
    return w, -*w-gL*np.sin()+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 30')
plt.legend()

#V = 50
state0 = [1,0]  # [theta0, omega0]
 = 0.1
gL = 1
A = 2
v = 50
t = np.arange(0,100,0.1)
def f(state, t):
    , w = state
    return w, -*w-gL*np.sin()+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 30')
plt.legend()
```
