```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#
def logisticmap(r,n,f):
    i = 1
    F = [f]
    while(i<=n):
        f = r*f*(1-f)
        i += 1
        F += [f]
    return F


#r = 0.5
F = logisticmap(r=0.5,n=100,f=0.5) 
plt.plot(F,label="$r=0.5$")
plt.ylabel('Fn')
plt.xlabel('n')

#r = 0.95
F = logisticmap(r=0.95,n=100,f=0.5) 
plt.plot(F,label="$r=0.95$")
plt.ylabel('Fn')
plt.xlabel('n')

#r = 2
F = logisticmap(r=2,n=100,f=0.5) 
plt.plot(F,label="$r=2$")
plt.ylabel('Fn')
plt.xlabel('n')

#r = 3.2
F = logisticmap(r=3.2,n=100,f=0.5) 
plt.plot(F,label="$r=3.2$")
plt.ylabel('Fn')
plt.xlabel('n')

#r = 3.8
F = logisticmap(r=3.8,n=100,f=0.5) 
plt.plot(F,label="$r=3.8$")
plt.ylabel('Fn')
plt.xlabel('n')

#r = 3.8
F = logisticmap(r=3.8,n=100,f=0.501) 
plt.plot(F,label="$r=3.8 f=0.501$")
plt.ylabel('Fn')
plt.xlabel('n')

plt.grid()
plt.legend();
```
