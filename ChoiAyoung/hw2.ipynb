$$ f_{n+1} = rf_{n}(1-f_{n}) $$

```python
r = 4.5
x = 0.5
n = 1

def f(x):
    x = r*x*(1-x)
    return x

while (n<=20):
    x=r*x*(1-x)
    n=n+1
    print(x)
```

$$  F_{n}=F_{n-1}+F_{n-2} $$

```python
def F(n):
    if 0<n<=2:
        n=1
    elif 2<n:
        n=F(n-1)+F(n-2)
    else:
        n=print('값이 잘못 입력되었습니다. 재실행하세요.')
    return n

n=int(input('n값 : '))

print(F(n))
```
