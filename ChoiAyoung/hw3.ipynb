## 1. Leibniz Pi formula

```python
def f(n):
    x = 0
    for i in range(1,n+1):
        x=((-1)**(i-1)/(2*i-1))+x
    return x
    
print(f(n=int(input('n값:'))))

import numpy as np
import matplotlib.pyplot as plt

A=np.arange(1,100,1)
a=[f(n) for n in A]
b=[np.pi/4 for n in A]
plt.plot(a,color='pink',label='Leibniz formula $ \pi $')
plt.plot(b,color='skyblue',label='$ \pi $/4')
plt.xlabel(r'$ n $', fontsize=10)
plt.legend()
```

## 2. Logistic map

## 3. sinc function

```python
a=np.arange(-20,21,0.1)
b=[np.sin(x)/x for x in a]
plt.plot(a,b,color='skyblue')
```
