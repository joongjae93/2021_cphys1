```python
import numpy as np
import matplotlib.pyplot as plt
```

### 만약 주기적인 힘이 $ F = Ce^{iwt} $ 로 주어지면 x는 다음의 $ x_i(t) = a_ie^{iwt} $ 의 해를 가진다.
$$ Ma=b $$
### m=1, k=1, C=1, N=40 고정

```python
k=1
P=-np.eye(40,40,1)
B1=np.eye(40,40,41)
B2=np.eye(40,40,41)
B1[0,0]=-k
B2[39,39]=-k
b=np.eye(40,1)
```

```python
def B(d):
    return d*np.identity(40)+B1+B2
def M(d):
    return np.mat(B(d)+P+np.transpose(P))
def a(d):
    return np.linalg.solve(M(d),b)
def aT(d):
    return np.transpose(a(d))
```

```python
d=0.5
print(np.linalg.det(M(d)))
i = np.arange(0,39)
plt.plot(i,aT(d)[0,i],'.-',color='red',label='$Ma=b$')
plt.xlabel('i', size = 20)
plt.ylabel('$a_i$', size = 20)
plt.legend()
plt.title('d=-0')
```

### $d=1$ 인 경우, $a_{3n}=0$이고, $a_{4n-3}=a_{4n-2}=-a_{4n-1}=-a_{4n}$으로 $a_{3n}$을 마디로 하는 정상파로 볼 수 있다. <br>또한 이런 해석은 system의 고유진동수 $w_0=\sqrt(k/m)=1$<br>$d=2k-mw^2=2-w^2 ⇒ w = 1, w=w_0$ 인 것과 의미가 통한다.<br> $w=w_0$ 일 때 공명현상이 안 일어나는 건 위상차이가 있기 때문이라 보이는데, $d=0.908$ 근처에서 진폭이 굉장히 증가하기 때문에 위상차이가 $Δd=1.0-0.908$ 만큼 난다고 추측할 수 있다.<br>반면 $d<0 (M^T≠0)$ 일 때는 d가 커질수록 w가 커진다. 즉 힘의 크기가 짧은 주기로 바뀌게 되는데, 이렇게 되면<br>힘을 받은 첫번째 물체 $a_1$가 뒤의 $a_i (i>1)$ 물체에게 힘을 전달하기 전에 F의 방향이 바뀌어<br>방금 전달하려는 힘을 상쇄 시켜버리기 때문에 결국 $d$ 가 커질 수록 직접 힘을 받는 $a_1$ 만 진폭을 갖고<br> $a_i (i>0)$ 인 물체들은 진동하지 않게 된다.<br> $w=2w_0$일 때 $d=-3$가 되며 $a_i>a_{i+1}$관계가 명확히 보이기 시작하는데, 이게 심화되며 결국 $d=-100$ 같은 상황이 됐다고 보인다.

```python
d=np.eye(40,1,-39)
c=b+d
def A(d):
    return np.linalg.solve(M(d),c)
def AT(d):
    return np.transpose(A(d))
```

```python
d=0.5
print(np.linalg.det(M(d)))
i = np.arange(0,39)
plt.plot(i,AT(d)[0,i],'o-',color='blue',label='$Ma=c$')
plt.plot(i,aT(d)[0,i],'.-',color='red',label='$Ma=b$')
plt.xlabel('i', size = 20)
plt.ylabel('$a_i$', size = 20)
plt.legend()
plt.title('d=0')
```

##### 첫번째 물체만 힘을 받던 $Ma=b$ 그래프도 참고를 위해 점 그래프 형태로 추가했다.<br> 
### 첫번째 물체와 마지막 물체가 동일한 크기의 힘을 받는 중이므로 $Ma=b$ 그래프 보다<br>$Ma=c$ 그래프가 좌우대칭인 경향이 보인다.<br>또한 한 쪽에서만 힘이 올 때는 볼 수 없었던 양 끝에서 온 힘이 증폭되는 것을  $d=-1.9753767$ 에서 볼 수 있다.<br> 양 끝에서 힘이 와도 고유진동수는 변함없기 때문에 여전히 $d=1$에서 정상파 형태를 갖고,<br> $w=2w_0$ 가 되는 $d=-3$ 에서 양 끝에서 시작된 두 힘이 서로 가운데 지점에서 상쇄되는 것을 볼 수 있다.<br> 그런데 $d$가 작아질수록 좌우대칭 꼴이 상하대칭 꼴로 바뀔 걸로 예상했는데, $d$가 작아질 수록 $Ma=b$ 형태와 같아진다. $w=w_0 (d=1)$에서 공명현상이 안 보인 것 처럼 numpy 명령어 중에 임의의 위상차를 주는 요소가 있어서 이런 형태가 된 게 아닐까 의심된다.
