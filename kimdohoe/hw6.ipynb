```python
import numpy as np
import matplotlib.pyplot as plt

u = [1.0]
w = 0.0
t = [0.0]
dt = 0.01

for i in range(0, 1000):
    w += -np.sin(u[i]) * dt
    u.append(u[i] + w * dt)
    t.append(t[i] + dt)

plt.title('θ(0) = 1.0, w(0) = 0.0')
plt.xlabel('t(s)')
plt.ylabel('θ(r)')
plt.plot(t, u)
```

```python
u0 = 0.0
w0 = 0.0

for u0 in np.arange(1.0, 2.0, 0.5):
    for w0 in np.arange(0.0, 2.0, 1.0):
        u = [u0]
        w = w0
        t = [0.0]
        dt = 0.01

        for i in range(0, 1000):
            w += -u[i] * dt
            u.append(u[i] + w * dt)
            t.append(t[i] + dt)

        plt.title("""
        blue  : θ(0) = 1.0, w(0) = 0.0 
        yellow: θ(0) = 1.0, w(0) = 1.0 
        green : θ(0) = 1.5, w(0) = 0.0 
        red   : θ(0) = 1.5, w(0) = 1.0 
        """)
        plt.xlabel('t(s)')
        plt.ylabel('θ(r)')
        plt.plot(t, u)
```

```python
u = [1.0]
w = 0.0
t = [0.0]
dt = 0.01
b = 0.1

for i in range(0, 4000):
    w += (-b * w - np.sin(u[i])) * dt
    u.append(u[i] + w * dt)
    t.append(t[i] + dt)

plt.title('θ(0) = 1.0, w(0) = 0.0')
plt.xlabel('t(s)')
plt.ylabel('θ(r)')
plt.plot(t, u)
```

```python
for v in np.arange(1, 1.5, 0.2):
    u = [-1.0]
    w = 0.0
    t = [0.0]
    dt = 0.01
    b = 0.1
    A = 0.1

    for i in range(0, 10000):
        w += (A * np.cos(v * t[i]) - b * w - np.sin(u[i])) * dt
        u.append(u[i] + w * dt)
        t.append(t[i] + dt)

    plt.title("""
        θ(0) = -1.0, w(0) = 0.0, A = 0.1
        blue  : v = 1.0 
        yellow: v = 1.2 
        green : v = 1.4 
        """)
    plt.xlabel('t(s)')
    plt.ylabel('θ(r)')
    plt.plot(t, u)
```

## 외력의 각진동수도 v = 1 로 두어 외력과 진자운동의 위상을 일치시킨 경우에는 
## 외력이 운동방향으로 작용하여 damping 이 일어나지 않는다
## v 를 1.2, 1.4 로 증가시켜 진자운동의 각진동수와 위상차이를 크게 할 수록
## 외력과 운동의 방향이 불일치하는 시간이 길어지므로 damping 이 빠르게 일어난다. ##
