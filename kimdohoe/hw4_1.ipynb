```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
import math 
```

```python
def y(t):
    return (2/(np.pi)**(1/2))*np.exp(-(t**2))

def trapez(f,a,b,n):
    s, t, dt = 0, a, (b-a)/n
    while(t<b):
        s += dt*(0.5*f(t)+0.5*f(t+dt))        
        t=t+dt
    return s

trapez(y,0,100,100)
```

```python
h=[j for j in range(-20,20)]
g= [trapez(y, 0, x, 100) for x in range(-20,20)]
plt.plot(h,g)
```

```python
trapez(y, 0, 1, 100)
```

```python
def erf(x,n):
    f=0
    for i in range(n+1):
        f =f+((-1)**i)*(x**(2*i+1))/(2*i+1)*(math.factorial(i))
    return f
```

```python
x = np.arange(-20,20,0.1)
plt.subplot(311)
plt.plot(x, erf(x,1), color='blue')
```

```python
plt.subplot(311)
plt.plot(x, erf(x,2), color='blue')
```
