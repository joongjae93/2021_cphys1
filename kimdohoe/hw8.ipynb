```python
import random
import numpy as np
```

```python
#3차원
def s(n):
    count=0
    for i in range(n):
        x,y,z=random.random(),random.random(),random.random()
        if x**2+y**2+z**2<1:
            count+=1
    return (2**3*count)/n
```

```python
s(10000)
```

```python
#4차원
def s(n):
    count=0
    for i in range(n):
        x,y,z,j=random.random(),random.random(),random.random(),random.random()
        if x**2+y**2+z**2+j**2<1:
            count+=1
    return (2**4*count)/n
```

```python
s(10000)
```

```python
#5차원
def s(n):
    count=0
    for i in range(n):
        x,y,z,j,k=random.random(),random.random(),random.random(),random.random(),random.random()
        if x**2+y**2+z**2+j**2+k**2<1:
            count+=1
    return (2**5*count)/n
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ave1(b1,n):
    states=np.ones(n)
    for t in range(n*100):
        i=random.randint(0,n-1)
        DE=-2*states[i]
        if DE<0:
            states[i]=-1*states[i]
        elif random.random()<np.exp(-b1*DE):
            states[i]=-1*states[i]
    return sum(states)
```

```python
def ave2(b2,n):
    states=np.ones(n)
    for t in range(n*100):
        i=random.randint(0,n-1)
        DE=-2*states[i]
        if DE<0:
            states[i]=-1*states[i]
        elif random.random()<np.exp(-b2*DE):
            states[i]=-1*states[i]
    return sum(states)
```

```python
def ave3(b3,n):
    states=np.ones(n)
    for t in range(n*100):
        i=random.randint(0,n-1)
        DE=-2*states[i]
        if DE<0:
            states[i]=-1*states[i]
        elif random.random()<np.exp(-b3*DE):
            states[i]=-1*states[i]
    return sum(states)
```

```python
temp1=np.arange(1,20,2)
y1=[ave1(1/t,1000) for t in temp1]
b1=np.arange(1,20,0.1)
th1=-1000*np.tanh(1/b1)

temp2=np.arange(1,20,2)
y2=[ave2(1/t,1000) for t in temp2]
b2=np.arange(1,20,0.1)
th2=-1000*np.tanh(1/b2)

temp3=np.arange(1,20,2)
y3=[ave3(1/t,1000) for t in temp3]
b3=np.arange(1,20.1)
th3=-1000*np.tanh(1/b3)
```

```python
plt.xlabel('temperature')
plt.ylabel('<E>')
plt.plot(temp1,y1,'o',color='black')
plt.plot(b1,th1);
plt.plot(temp2,y2,'o',color='red')
plt.plot(b2,th2);
plt.plot(temp3,y1,'o',color='brown')
plt.plot(b3,th3);
```
