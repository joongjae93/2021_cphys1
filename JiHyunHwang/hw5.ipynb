```python
import numpy as np
import matplotlib.pyplot as plt

def D (k,m,w) :
    d = 2*k-m*w**2
    return d

def Matrix(n,d,k) :
    dia1 = (-k)*np.ones((n-1),dtype="float64")
    dia2 = (-k)*np.ones((n-1),dtype="float64")
    dia = d*np.ones((n,),dtype = "float64")
    dia[0] = d-k
    dia[n-1] = d-k
    M = np.zeros((n,n),dtype = 'float64')+np.diag(dia1,k=-1)+np.diag(dia)+ np.diag(dia2,k=1)
    return M
def Matrixb(n,C) :
    b = np.zeros((n,1))
    b[0] = C
    b[-1] = C
    return b


x = np.linalg.solve(Matrix(40,D(1,1,1),1),Matrixb(40,1))
x1 = np.linalg.solve(Matrix(40,D(1,1,2),1),Matrixb(40,1))
x2 = np.linalg.solve(Matrix(40,D(1,1,3),1),Matrixb(40,1))

plt.plot(range(0,40,1),x,label = "w = 1")
plt.plot(range(0,40,1),x1, label = "w = 2")
plt.plot(range(0,40,1),x2, label = "w = 3")
plt.xlabel("N")
plt.ylabel("$a_i$")
plt.legend()
```
