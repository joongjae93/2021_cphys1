# 1. Pendulum

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math

t = np.arange(0,500,0.01)
state0 = [1,0.1]

def f(state,t) :
    s,w = state
    return w, -1*np.sin(np.radians(s))

states = odeint(f,state0, t)
plt.plot(t,states[:,0])
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Pendulum Oscilation")
```

# 3. Damped Pendulum

```python
t = np.arange(0,100,0.01)
state0 = [1,0]
B = 0.1

def f(state,t) :
    s,w = state
    return w, -B*w-1*np.sin(s)

states = odeint(f,state0, t)
plt.plot(t,states[:,0])
plt.plot([0,100],[0,0],"--")
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
```
