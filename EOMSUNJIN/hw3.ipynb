### Leibniz Pi formula

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import math
```

```python
def Leibniz_Pi_formula(n):
    x, y = [], []
    f = 0
    l = [i for i in range(1, n + 1) if i % 2 == 1]
    for i in range(0, len(l)):
        if i % 2 == 0:
            f += 1 / l[i]
            y.append(f)
            x.append(l[i])
        elif i % 2 == 1:
            f += 1 / -l[i]
            y.append(f)
            x.append(l[i])
    plt.plot(x, y, '-b', label='Leibniz formula for pi)
    plt.hlines(math.pi / 4, 0, 100, color='r', label=r'$\pi$')
    plt.legend()
    plt.show()
    
```

```python
Leibniz_Pi_formula(100)
```

```python
def logistic_map(r, f, n,n2):
    y = []
    x = []
    for i in range(1, n + 1):
        f = r * f * (1 - f)
        y.append(f)
        x.append(i)
    plt.plot(x,y)
    plt.subplot(n2)
```

```python
logistic_map(0.5,0.5,100,231)
logistic_map(0.95,0.5,100,232)
logistic_map(2,0.5,100,233)
logistic_map(3.2,0.5,100,234)
logistic_map(3.8,0.5,100,235)
logistic_map(3.8,0.51,100,236)
```

```python
def sinc(n):
    x = np.arange(-10,10,0.1)
    y = []
    for i in x:
        s = math.sin(i) / i
        y.append(s)
    plt.plot(x,y)
```

```python
sinc(10)
```

```python
def sinc_talyor(x,n):
    f=0
    for i in range(n+1):
        f=f+((-1)**i)*(x**(2*i))/math.factorial((2*i)+1)
    return f
    
```

```python
x = np.arange(-10,10,0.1)
plt.plot(x,sinc_talyor(x,1), color='green')
plt.plot(x,sinc_talyor(x,4), color='red')
plt.plot(x,sinc_talyor(x,9), color='blue')
plt.plot(x,sinc_talyor(x,16), color='black')
plt.axis([-10,10,-1,1])
```
