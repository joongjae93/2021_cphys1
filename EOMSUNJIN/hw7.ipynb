# Pauli Matrices

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sx=np.array([[0,1],[1,0]])
ex=np.linalg.eig(sx)

sy=np.array([[0,-1j],[1j,0]])
ey=np.linalg.eig(sy)

sz=np.array([[1,0],[0,-1]])
ez=np.linalg.eig(sz)
```

```python
ex,ey,ez
```
