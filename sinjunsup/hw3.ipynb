```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def le(n):
    i=0
    f=0
    while(i<n):
        i= i+1
        f= f+1/((2*i-1)*(-1)**(i+1))
    return f
```

```python
n=100
x=np.arange(0,100,2)
plt.plot(x,[le(i+1) for i in range(50)],label='Leibniz formula for π')
plt.plot(x,[np.pi/4 for i in range(50)],label='π/4')
plt.xlabel("n")
plt.ylim(0.65,1)
plt.legend()
```

## logistic map

```python
def lo(r,n):
    f=0.501
    i=0
    while(i<n):
        f=r*f*(1-f)
        i=i+1
    return f
    
```

```python
lo(0.5,4)
```

```python
#r=0.5, f0=0.5
n=100
x=np.arange(1,100,1)
plt.plot(x,[lo(0.5,i) for i in range(99)])
plt.xlabel("n")
plt.ylabel("fn",rotation=0)
```

## sinc function

```python
x=np.arange(-20,20,0.1)
plt.plot(x,np.sin(x)/x)
```

```python
def s(x,n):
    i=0
    fac=1
    f=0
    while(i<n):
        r=i
        for k in range(1,2*r+2):
            fac =fac*k
        f=f+((-1)**i*x**(2*i))/fac 
        i=i+1
    return f
```

```python
x=np.arange(-20,20,0.1)
plt.plot(x,s(x,10),label='talyer') #n=5
y=np.arange(-20,20,0.1)
plt.plot(y,np.sin(y)/y,label='sin(y)/y')
plt.ylim(-1,1)
plt.legend()
```
