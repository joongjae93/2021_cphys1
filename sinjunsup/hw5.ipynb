```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M=np.zeros((40,40))
b1=np.zeros((40,1))
b1[0,0]=1
b2=np.zeros((40,1))
b2[0,0],b2[39,0]=1,1
```

```python
def ieg(w): # 행렬 M에값을 대입하는 함수
    i=0
    d=2-w**2
    for i in range(0,39):
        M[i+1,i]=-1
        M[i,i+1]=-1
        M[i,i]=d
    M[0,0],M[39,39]= d-1,d-1
```

```python
ieg(1) #w =1 
a=np.linalg.solve(M,b1)
A=np.linalg.solve(M,b2)
x = np.arange(0,40,1)
plt.plot(x,a,label='b1')
plt.plot(x,A,label='b2')
plt.legend()
```

```python
ieg(2) # w= 2
a=np.linalg.solve(M,b1)
A=np.linalg.solve(M,b2)
x = np.arange(0,40,1)
plt.plot(x,a,label='b1')
plt.plot(x,A,label='b2')
plt.legend()
```
