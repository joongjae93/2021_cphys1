# 1. Leibniz Pi formula

\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots 
\end{align}
이 식은 라이프니츠의 파이 공식이다.  
1. 다음의 식을 구현하는 함수를 만들어라.  
\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}
2. 이 식을 이용하여 아래처럼 그림을 그려라.  

```python
# 1번 문제
# n은 홀수여야 함
# n이 짝수일 시 n 직전의 홀수까지 공식 계산
def leibniz(n):
    if n <= 0: 
        return None
    else:
        fn = [1/(2 * i - 1) * ((-1) ** (i - 1)) for i in range(1, n // 2 + 2)]
        return sum(fn)
```

```python
# 2번 문제
# x는 1~100 사이의 홀수의 배열
# y는 x의 원소들을 입력값으로 하는 함숫값
# pi_value는 π/4를 그래프로 그리기 위한 함숫값 배열
x = np.array([i for i in range(1, 101, 2)])
y = np.array([leibniz(i) for i in x])
pi_value = np.array([np.pi/4 for i in x])
plt.xlabel('n')
plt.plot(x, y)
plt.plot(x, pi_value)
plt.legend(['Leibniz formular for π', 'π/4'])

plt.show()
```

# 2. Logistic Map

로지스틱 맵은 다음처럼 주어진다.
$$ f_{n+1} = r f_n (1- f_n). $$  

1. $f_n$을 구하는 함수를 작성하여라.
2. 작성한 함수를 이용하여 다음의 조건에서 $n=1$에서 $100$까지 로지스틱 맵의 값을 구하고 그래프로 그려라. 즉, x축은 $n$, y축은 $f_n$인 그래프를 그려라.
  * $r=0.5$, $f_0$=0.5
  * $r=0.95$, $f_0$=0.5
  * $r=2$, $f_0$=0.5
  * $r=3.2$, $f_0$=0.5
  * $r=3.8$, $f_0$=0.5
  * $r=3.8$, $f_0$=0.501

```python
# 1번 문제
def logistic_map(r, f0, n):
    for i in range(n):
        f0 = r * f0 * (1 - f0)
    return f0
```

```python
# 2번 문제 
# 그래프 작성 함수
def draw_graph(r, f0):
    x = np.array([i for i in range(1, 101)])
    y = np.array([logistic_map(r, f0, i) for i in x])
    plt.plot(x, y)
    plt.xlabel('n')
    plt.ylabel('f_n')
    plt.show()
```

```python
# 2번 문제
# r = 0.5, f0 = 0.5
draw_graph(0.5, 0.5)
```

```python
# 2번 문제
# r = 0.95, f0 = 0.5
draw_graph(0.95, 0.5)
```

```python
# 2번 문제
# r = 2, f0 = 0.5
draw_graph(2., 0.5)
```

```python
# 2번 문제
# r = 3.2, f0 = 0.5
draw_graph(3.2, 0.5)
```

```python
# 2번 문제
# r = 3.8, f0 = 0.5
draw_graph(3.8, 0.5)
```

```python
# 2번 문제
# r = 3.8, f0 = 0.501
draw_graph(3.8, 0.501)
```

# 3. sinc function

1. 싱크 함수를 그려보아라. 싱크함수는 다음처럼 주어진다.
$$ sinc(x)  = \frac{\sin x}{x}. $$
2. 싱크 함수를 테일러 전개하면 다음과 같다.
$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$ 
위의 식의 우변을 $n$을 바꾸어 가면서  그래프로 그리고 싱크함수와 비교하여라. 

```python
import math
# 1번 문제
# 싱크 함수 정의
def sinc(x):
    if x == 0: 
        return 1
    else:
        return math.sin(x)/x

# 그래프 작성
x = np.array([i for i in range(1, 101)])
y = np.array([sinc(i) for i in x])
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.show()
```

```python
import math
# 2번 문제
# 테일러 전개식 함수
def sinc_taylor(n, x):
    op = [((-1) ** i) * (x ** (2 * i)) / math.factorial(2 * i + 1) for i in range(0, n+1)]
    return sum(op)

# 그래프 작성 함수
def make_sinc_and_taylor_graph(n):
    x = np.array([i for i in range(1, 101)])
    sinc_y = np.array([sinc(i) for i in x])
    taylor_y = np.array([sinc_taylor(n, i) for i in x])
    plt.plot(x, sinc_y)
    plt.plot(x, taylor_y)
    plt.xlabel('x')
    plt.legend(['sinc', 'taylor of sinc'])
    plt.show()
    
```

```python
# 2번 문제
# n = 1
make_sinc_and_taylor_graph(1)
```

```python
# 2번 문제
# n = 10
make_sinc_and_taylor_graph(10)
```

```python
# 2번 문제
# n = 50
make_sinc_and_taylor_graph(50)
```

```python
# 2번 문제
# n = 100
make_sinc_and_taylor_graph(100)
```
