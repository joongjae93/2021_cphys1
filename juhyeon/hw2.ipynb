## Logistic Map (part 2)

로지스택 맵은 다음처럼 주어진다.

$$ f_{n+1} = rf_n(1-f_n). $$

주어진 $r$과 초기조건 $f_0$에 $f_n$을 다음의 과정을 통해 구하여라.

1. $f_n$을 구하는 함수를 작성하여라.
2. 작성한 함수를 이용하여 다음의 조건에서 $n = 1$에 $20$까지 로지스틱 맵의 값을 구하여라.

 * $ r = 4.5, f_0 = 0.5 $
 * $ r = 4.5, f_0 = 0.51 $
 * $ r = 4.5, f_0 = 0.501 $
 3. 초기조건에 민감함을 확인하여라.

```python
def logistic(r,f):
    for n in range(20):
        f = r*f*(1-f)
        print('f', n,'=',f )
```

```python
logistic(4.5,0.5)
```

```python
logistic(4.5,0.51)
```

```python
logistic(4.5,0.501)
```

## 피보나치 수 (Fibonacci numbers)
피보나치 수(Fibonacci numbers)는 처음 두 개의 항이 1이며 그 뒤의 모든 항은 앞의 두 개의 항의 합으로 주어지는 수열이다. 따라서 다음처럼 주어지는 수열이다.
$$1, 1, 2, 3, 5, 8, ...$$
$n$번째 피보나치 수 $F_n$은 다음의 식으로 구할 수 있다.
$F_1 = F_2 = 1, F_n = F_{n-1}+F_{n-2},$

$n$번째 피보나치 수를 구하는 함수를 프로그래밍하여라.

```python
def fibo(n) :
    a,b = 1,1
    if  n == 1 or n == 2:
        return 1
    for i in range(1,n):
        a,b = b, a+b
    return a
```

```python
for n in range(20):
    n = n+1
    print(fibo(n))
```
