## 1. Leibniz Pi formula

$$ \frac{\pi}{4} = \frac{1}{1} - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} - \cdots$$

이 식은 라이프니츠의 파이 공식이다.

1. 다음의 식을 구현하는 함수를 만들어라.

$$  f(n) = \frac{1}{1} - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.         $$ 

```python
def lei(n):
    i,f = 1, 0        
    while(2*i-1<=n):
        f = f + (-1)**(i+1)/(2*i-1)
        i += 1
    return f

lei(100)
```

2. 이 식을 이용하여 아래처럼 그림을 그려라.

```python
def lei(n):
    i,f = 1, 0
    f_list = []
    while(2*i-1<=n):
        f = f + (-1)**(i+1)/(2*i-1)
        i += 1
        f_list += [f]
    return f_list

f_list = lei(100)
plt.plot(f_list, label='Leibniz formula for $\pi$' )
plt.axhline(y=math.pi/4, xmin = 0.05, xmax = 50,color='sandybrown', label='$\pi/4$')
plt.xlabel('n')
plt.legend();
```

## 2. Logistic Map

로지스틱 맵은 다음처럼 주어진다.

$$  f_{n+1} = rf_n(1-f_n). $$

1. $f_n$을 구하는 함수를 작성하여라.
2. 작성한 함수를 이용하여 다음 조건 $n = 1$에서 100까지 로지스틱 맵의 값을 구하고 그래프로 그려라.  <br/>즉, x축은 $n$, y축은 $f_n$인 그래프를 그려라.

1. $ r = 0.5, f_0 = 0.5 $  
2. $ r = 0.95, f_0 = 0.5 $   
3. $ r = 2, f_0 = 0.5 $  
4. $ r = 3.2, f_0 = 0.5 $  
5. $ r = 3.8, f_0 = 0.5 $  
6. $ r = 3.8, f_0 = 0.501 $  

```python
def logistic(r,n,f0):
    f = f0
    f_list = []
    for n in range(n):
        f = r*f*(1-f)
        f_list += [f]
    return f_list
```

```python
# 1. r = 0.5, f0 = 0.5
print('r = 0.5, f0 = 0.5')
print(logistic(0.5,100,0.5))

# 2. r = 0.95, f0 = 0.5
print('r = 0.95, f0 = 0.5')
print(logistic(0.95,100,0.5))

# 3. r = 2, f0 = 0.5
print('r = 2, f0 = 0.5')
print(logistic(2,100,0.5))

# 4. r = 3.2, f0 = 0.5
print('r = 3.2, f0 = 0.5')
print(logistic(0.5,100,0.5))

# 5. r = 3.8, f0 = 0.5
print('r = 3.8, f0 = 0.5')
print(logistic(3.8,100,0.5))

# 6. r = 3.8, f0 = 0.501
print('r = 3.8, f0 = 0.501')
print(logistic(3.8,100,0.501))
```

## 3. sinc function

1. 싱크 함수를 그려보아라. 싱크 함수는 다음처럼 주어진다.
$$  sinc(x) = \frac{sinx}{x} . $$
2. 싱크 함수를 테일러 전개하면 다음과 같다.
$$ sinc(x) = \frac{sinx}{x} = \Sigma^\infty_{n = 0} \frac{(-1)^{n}x^{2n}}{(2n+1)!}. $$
위의 식의 우변 n을 바꾸어 가면서 그래프로 그리고 싱크함수와 비교하여라.

```python
# 1. 싱크 함수를 그려보아라.

a = np.arange(-20,20,0.1)
b = [np.sin(x)/x for x in a]
plt.plot(a,b);
```

```python
# 2. 위의 식의 우변 n을 바꾸어 가면서 그래프로 그리고 싱크함수와 비교하여라.

def sinct (x,n):
    i,f = 0, 0
    for i in range(n+1):
        f = f + ((-1)**i*x**(2*i))/math.factorial(2*i+1)
    return f

x = np.arange(-20,20,0.1)
fig = plt.figure(figsize=(16,3))
subplot = fig.add_subplot(1,3,1)
plt.plot(a,b)
plt.title('$sin(x)/x$')
plt.xlabel('$x$')
plt.ylabel('$sinc(x)$')

subplot = fig.add_subplot(1,3,2)
plt.plot(x,sinct(x,0),label='n=0')
plt.plot(x,sinct(x,1),label='n=1')
plt.plot(x,sinct(x,2),label='n=2')
plt.plot(x,sinct(x,3),label='n=3')
plt.plot(x,sinct(x,4),label='n=4')
plt.title('Taylor expansion')
plt.xlabel('$x$')
plt.ylim([-0.3, 1.1])
plt.legend()

subplot = fig.add_subplot(1,3,3)
plt.plot(x,sinct(x,10),label='n=10')
plt.plot(x,sinct(x,20),label='n=20')
plt.plot(x,sinct(x,30),label='n=30')
plt.plot(x,sinct(x,40),label='n=40')
plt.plot(x,sinct(x,50),label='n=50')
plt.title('Taylor expansion')
plt.xlabel('$x$')
plt.ylim([-0.3, 1.1])
plt.legend();
```

$n$ 값이 커질수록 싱크함수($sinc(x) = \frac{sin(x)}{x}$)에 가까워 지는 것을 볼 수 있다.
