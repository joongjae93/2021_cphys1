## Logistic Map

로지스택 맵은 다음처럼 주어진다.

$$ f_{n+1} = rf_n(1-f_n). $$

주어진 $r$과 초기조건 $f_0$에 대해, 반복문(while or for)을 이용해서 n이 1~10일 때, $f_n$을 구하여라.

 1. $ r = 0.5, f_0 = 0.5 $
 2. $ r = 2.5, f_0 = 0.5 $
 3. $ r = 4.5, f_0 = 0.5 $
 4. $ r = 4.5, f_0 = 0.51 $

```python
# 1. r = 0.5, f_0 = 0.5

r = 0.5
f = 0.5
for n in range(10):
    n = n+1
    print(n)
    f = r*f*(1-f)
    print(f)
    print("f",n,"=",f)
```

```python
# 2. r = 2.5, f_0 = 0.5

r = 2.5
f = 0.5
for n in range(10):
    n = n+1
    f = r*f*(1-f)
    print("f",n,"=",f)
```

```python
# 3. r = 4.5, f_0 = 0.5

r = 4.5
f = 0.5
for n in range(10):
    n = n+1
    f = r*f*(1-f)
    print("f",n,"=",f)
```

```python
# 4. r = 4.5, f_0 = 0.51

r = 4.5
f = 0.51
for n in range(10):
    n = n+1
    f = r*f*(1-f)
    print("f",n,"=",f)
```
