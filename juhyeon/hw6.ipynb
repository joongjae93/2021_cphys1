## 1. Pendulum
다음처럼 주어지는 진자의 운동방정식을 풀어라.$$ \frac{d^2 \theta}{dt^2}  + \frac{g}{L} \sin \theta = 0. $$$\beta$는 마찰의 세기, $L$은 진자의 길이, $g$는 중력가속도이다. 흔히 작은 $\theta$를 가정하고 $\sin\theta \approx \theta$ 근사를 이용하여 해석적인 해를 구한다. 여기에서는 $\sin \theta$를 직접 이용하여 수치해를 구해보자.

2차 미분방정식을 2개의 1차 미분방정식으로 분리하자.$$ \frac{d\theta}{dt} = w, \\
\frac{d w}{dt} = - \frac{g}{L} \sin \theta . $$

$\frac{g}{L}=1$을 가정하고 운동의 해를 구해보자. scipy.integrate.odeint를 포함하여 어떤 방법을 이용하여도 된다.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

plt.figure(figsize=(16,8));
plt.subplot(2,1,1)
plt.title('theta = 0.1')
state0 = [0.1,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*np.sin(theta)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'w = 0')
plt.xlabel('t')
plt.legend()


state0 = [0.1,0.5]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*np.sin(theta)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label ='w = 0.5')
plt.legend()

state0 = [0.1,1]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*np.sin(theta)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0], label = 'w = 1')
plt.legend()

state0 = [0.1,1.5]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*np.sin(theta)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0], label = 'w = 1.5')
plt.legend()

plt.subplot(2,1,2)
plt.title('w = 0')
state0 = [0.1,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*np.sin(theta)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'theta = 0.1')
plt.xlabel('t')
plt.legend()


state0 = [0.2,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*np.sin(theta)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label ='theta = 0.2')
plt.legend()

state0 = [0.3,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*np.sin(theta)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0], label = 'theta = 0.3')
plt.legend()

state0 = [0.5,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*np.sin(theta)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0], label = 'theta = 0.5')
plt.legend();
```

## 2. Pendulum (small oscillations)
1번 문제와 같은 진자에 대해 평형점 근처에서 조금씩 진동하고 있는 상황을 고려하자. 그러면
$\sin\theta \approx \theta$ 근사를 적용할 수 있다.

이 경우는 다음과 같은 2개의 1차 미분방정식을 얻는다.$$ \frac{d\theta}{dt} = w, \\
\frac{d w}{dt} =  - \frac{g}{L} \theta . $$

$\frac{g}{L}=1$을 가정하고 운동의 해를 구해보자. scipy.integrate.odeint를 포함하여 어떤 방법을 이용하여도 된다. 초기 조건 $\theta_o, w_o$를 바꾸어 가면서 1번 문제의 결과와 비교해보자.

```python
plt.figure(figsize=(16,8))
plt.subplot(2,1,1)
plt.title('theta = 0.1')
state0 = [0.1,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*theta

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'w = 0')
plt.xlabel('t')
plt.legend()


state0 = [0.1,0.5]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*theta

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label ='w = 0.5')
plt.legend()

state0 = [0.1,1]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*theta

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0], label = 'w = 1')
plt.legend()

state0 = [0.1,1.5]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*theta

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0], label = 'w = 1.5')
plt.legend()

plt.subplot(2,1,2)
plt.title('w = 0')
state0 = [0.1,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*theta

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'theta = 0.1')
plt.xlabel('t')
plt.legend()


state0 = [0.2,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*theta

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label ='theta = 0.2')
plt.legend()

state0 = [0.3,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*theta

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0], label = 'theta = 0.3')
plt.legend()

state0 = [0.5,0]  # [theta0, omega0]
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -gL*theta

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0], label = 'theta = 0.5')
plt.legend();
```

$sin\theta$일 때에는 $\theta$ 근사로 theta 값이 작아질 수록 그 두 값이 거의 같아지지만 컴퓨터로는 극한값을 계산하기 어려우므로 $\theta$를 통해서 위상차 없이 깔끔한 그래프를 나타낼 수 있게 된다.

## 3. Damped Pendulum
다음처럼 주어지는 damped 진자의 운동방정식을 풀어라.$$ \frac{d^2 \theta}{dt^2} + \beta \frac{d \theta}{dt} + \frac{g}{L} \sin \theta = 0. $$$\beta$는 마찰의 세기, $L$은 진자의 길이, $g$는 중력가속도이다.

우선 2차 미분방정식을 2개의 1차 미분방정식으로 분리해 보자.$$ \frac{d\theta}{dt} = w, \\
\frac{d w}{dt} = - \beta w - \frac{g}{L} \sin \theta . $$

초기 상태는 $\theta_0 = 1, w_0 =0$이며 $\beta=0.1, \frac{g}{L}=1$을 가정하고 운동의 해를 구해보자. scipy.integrate.odeint를 포함하여 어떤 방법을 이용하여도 된다.

```python
state0 = [1,0]  # [theta0, omega0]
beta = 0.1
gL = 1
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -beta*omega-gL*np.sin(theta)
states = odeint(f, state0, t)
plt.figure(figsize=(16,4))
plt.plot(t,states[:,0])
plt.title('Damped Pendulum')
plt.xlabel('t');
```

## 4. Forced Pendulum
다음처럼 주어지는 forced 진자의 운동방정식을 풀어라.$$ \frac{d^2 \theta}{dt^2} + \beta \frac{d \theta}{dt} + \frac{g}{L} \sin \theta = A \cos (v t). $$$\beta$는 마찰의 세기, $L$은 진자의 길이, $g$는 중력가속도, $A$는 외부에서 주어지는 힘의 진폭, $v$는 외부에서 주어지는 힘의 진동수이다.

forced 진자의 움직임을 구해보고 $v$의 효과에 대해 살펴보아라.

```python
state0 = [1,0]  # [theta0, omega0]
beta = 0.1
gL = 1

plt.figure(figsize=(16,4))
plt.title('Forced Pendulum')
A = 2
v = 5
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -beta*omega-gL*np.sin(theta)+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 5')
plt.xlabel('t')
plt.legend()

v = 10
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -beta*omega-gL*np.sin(theta)+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 10')
plt.xlabel('t')
plt.legend()

v = 15
t = np.arange(0,100,0.01)
def f(state, t):
    theta, omega = state
    return omega, -beta*omega-gL*np.sin(theta)+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 15')
plt.xlabel('t')
plt.legend();
```

v가 커질수록 단파장의 효과가 줄어든다.
