## 1. Vibration in a one-dimensional system

ÏßàÎüâÏù¥ $m$Ïù∏ $N$Í∞úÏùò ÎèôÏùºÌïú Î¨ºÏ≤¥Í∞Ä ÏùºÎ†¨Î°ú ÎèôÏùºÌïú Ïö©ÏàòÏ≤†Ïóê Îß§Îã¨Î†§ ÏûàÎã§. $i$Î≤àÏß∏ Î¨ºÏ≤¥Ïùò Ïö¥ÎèôÎ∞©Ï†ïÏãùÏùÄ Îã§ÏùåÏ≤òÎüº Ï£ºÏñ¥ÏßÑÎã§.

$$m \frac{d^2 x_i}{dt^2} = k(x_{i+1}-x_{i}) - k(x_i-x_{i-1})$$

ÌïúÏ™Ω ÎÅùÏóêÏÑú Ïó∞Í≤∞Îêú Ïö©ÏàòÏ≤†Ïóê Ï£ºÍ∏∞Ï†ÅÏù∏ Ìûò( ùêπ )ÏùÑ Í∞ÄÌïòÍ≥† ÏûàÎã§. Í∑∏Îü¨Î©¥ Ïñë ÎÅùÎã®Ïóê Î¨ºÏ≤¥Ïóê Ïö¥ÎèôÎ∞©Ï†ïÏãùÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$m \frac{d^2 x_1}{dt^2} = k(x_{2}-x_{1}) + F \\
m \frac{d^2 x_N}{dt^2} = -k(x_{N}-x_{N-1})$$

ÎßåÏïΩ Ï£ºÍ∏∞Ï†ÅÏù∏ ÌûòÏù¥ $F=C e^{i wt}$Î°ú Ï£ºÏñ¥ÏßÄÎ©¥ $x$Îäî Îã§ÏùåÏùò $x_i(t)=a_i e^{iwt}$Ìï¥Î•º Í∞ÄÏßÑÎã§. Ï†ïÎ¶¨ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ Î∞©Ï†ïÏãùÏùÑ ÏñªÏùÑ Ïàò ÏûàÎã§.

$$(d -k)a_1 -k a_2 = C, \\
d a_i - k a_{i-1}- k a_{i+1} = 0, \\
(d - k)a_N - k x_{N-1} =0$$

Ïó¨Í∏∞ÏóêÏÑú $d=2k-mw^2$Ïù¥Îã§.

ÏúÑÏùò Ïö¥ÎèôÎ∞©Ï†ïÏãùÏùÑ ÌñâÎ†¨Î°ú ÌëúÌòÑÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$Ma=b $$$$
M=
\begin{pmatrix}
d-k & -k & 0 & 0 & 0 & \cdots \\
-k & d & -k & 0 & 0 & \cdots \\
0 & -k & d & -k & 0 & \cdots \\
\cdots &\cdots &\cdots &\cdots &\cdots & \cdots  \\
0 & 0 & 0& 0& -k & d-k 
\end{pmatrix}, \quad
a=
\begin{pmatrix}
a_1 \\ 
a_2 \\ 
a_3 \\ 
\cdots \\ 
a_N  
\end{pmatrix}, \quad
b=
\begin{pmatrix}
C \\ 
0 \\ 
0 \\ 
\cdots \\ 
0 
\end{pmatrix}
$$

* $m=1$, $k=1$, $C=1$, $N=40$ÏúºÎ°ú Í≥†Ï†ïÌïòÏÑ∏Ïöî.
* $w$ (Ï¶â, d)Î•º Î∞îÍæ∏Ïñ¥Í∞ÄÎ©¥ÏÑú $a_i$Î•º Íµ¨ÌïòÏÑ∏Ïöî.
* Ïù¥Î•º Í∑∏ÎûòÌîÑÎ°ú ÌëúÌòÑÌïòÍ≥† ÏùòÎØ∏Î•º ÏÑ§Î™ÖÌïòÏÑ∏Ïöî. 

* 100X100 ÌñâÎ†¨ÏùÑ ÏßÅÏ†ë ÏûÖÎ†•ÌïòÎäî Í≤ÉÏùÑ Í∏àÏßÄÌïòÏßÄÎäî ÏïäÏßÄÎßå Í∑∏Î†áÍ≤å ÌïòÏßÄ ÏïäÍ∏∞Î•º Î∞îÎûçÎãàÎã§.
* np.zeros(n) Ìï®ÏàòÎäî ÌÅ¨Í∏∞Í∞Ä nÏù¥Î©∞ Î™®Îì† ÏÑ±Î∂ÑÏù¥ 0Ïù∏ Î∞∞Ïó¥ÏùÑ ÎßåÎìúÎäî Î™ÖÎ†πÏñ¥ÏûÖÎãàÎã§.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1
k = 1
c = 1
n = 40
b = np.zeros((n,1))
b[0] = c
M = np.zeros((40,40))

w = 1  # wÏùÑ Î∞îÍæ∏Ïñ¥ Í∞ÄÎ©¥ÏÑú a_i Íµ¨ÌïòÍ∏∞
d = 2*k - m*w**2

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 1 ÏùºÎïå a_i,',a)
xdata = np.arange(1,41,1)
plt.figure(figsize=(15,12))
plt.xlabel('$i$')
plt.ylabel('$a_i$')
plt.plot(xdata,a,'-',marker = 'o',label='w=1')
plt.legend()

w = 2  # wÏùÑ Î∞îÍæ∏Ïñ¥ Í∞ÄÎ©¥ÏÑú a_i Íµ¨ÌïòÍ∏∞
d = 2*k - m*w**2

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 2 ÏùºÎïå a_i,',a)
xdata = np.arange(1,41,1)
plt.plot(xdata,a,'-',marker = 'o',label='w=2')
plt.legend()

w = 3  # wÏùÑ Î∞îÍæ∏Ïñ¥ Í∞ÄÎ©¥ÏÑú a_i Íµ¨ÌïòÍ∏∞
d = 2*k - m*w**3

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 3 ÏùºÎïå a_i,',a)
xdata = np.arange(1,41,1)
plt.plot(xdata,a,'-',marker = 'o',label='w=3')
plt.legend()

w = 4  # wÏùÑ Î∞îÍæ∏Ïñ¥ Í∞ÄÎ©¥ÏÑú a_i Íµ¨ÌïòÍ∏∞
d = 2*k - m*w**4

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 4 ÏùºÎïå a_i,',a)
xdata = np.arange(1,41,1)
plt.plot(xdata,a,'-',marker = 'o',label='w=4')
plt.legend()

w = 5  # wÏùÑ Î∞îÍæ∏Ïñ¥ Í∞ÄÎ©¥ÏÑú a_i Íµ¨ÌïòÍ∏∞
d = 2*k - m*w**5

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 5 ÏùºÎïå a_i,',a)
xdata = np.arange(1,41,1)
plt.plot(xdata,a,'-',marker = 'o',label='w=5')
plt.legend();
```

## 2. ÏñëÏ™Ω ÎÅùÏóêÏÑú ÌûòÏùÑ Í∞ÄÌïòÎ©¥ Ïñ¥ÎñªÍ≤å ÎêòÎäîÍ∞Ä?

Ï¶â, bÌñâÎ†¨Ïù¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ Í≤ΩÏö∞Ïóê ÎåÄÌï¥ÏÑú ÏÇ¥Ìé¥Î≥¥ÏïÑÎùº.

$$b=

* m=1, k=1, C=1, N=40ÏúºÎ°ú Í≥†Ï†ïÌïòÏÑ∏Ïöî.
* w (Ï¶â, d)Î•º Î∞îÍæ∏Ïñ¥Í∞ÄÎ©¥ÏÑú $a_i$Î•º Íµ¨ÌïòÏÑ∏Ïöî.
