## 1. Vibration in a one-dimensional system

ì§ˆëŸ‰ì´ $m$ì¸ $N$ê°œì˜ ë™ì¼í•œ ë¬¼ì²´ê°€ ì¼ë ¬ë¡œ ë™ì¼í•œ ìš©ìˆ˜ì² ì— ë§¤ë‹¬ë ¤ ìˆë‹¤. $i$ë²ˆì§¸ ë¬¼ì²´ì˜ ìš´ë™ë°©ì •ì‹ì€ ë‹¤ìŒì²˜ëŸ¼ ì£¼ì–´ì§„ë‹¤.

$$m \frac{d^2 x_i}{dt^2} = k(x_{i+1}-x_{i}) - k(x_i-x_{i-1})$$

í•œìª½ ëì—ì„œ ì—°ê²°ëœ ìš©ìˆ˜ì² ì— ì£¼ê¸°ì ì¸ í˜( ğ¹ )ì„ ê°€í•˜ê³  ìˆë‹¤. ê·¸ëŸ¬ë©´ ì–‘ ëë‹¨ì— ë¬¼ì²´ì— ìš´ë™ë°©ì •ì‹ì€ ë‹¤ìŒê³¼ ê°™ë‹¤.

$$m \frac{d^2 x_1}{dt^2} = k(x_{2}-x_{1}) + F \\
m \frac{d^2 x_N}{dt^2} = -k(x_{N}-x_{N-1})$$

ë§Œì•½ ì£¼ê¸°ì ì¸ í˜ì´ $F=C e^{i wt}$ë¡œ ì£¼ì–´ì§€ë©´ $x$ëŠ” ë‹¤ìŒì˜ $x_i(t)=a_i e^{iwt}$í•´ë¥¼ ê°€ì§„ë‹¤. ì •ë¦¬í•˜ë©´ ë‹¤ìŒê³¼ ê°™ì€ ë°©ì •ì‹ì„ ì–»ì„ ìˆ˜ ìˆë‹¤.

$$(d -k)a_1 -k a_2 = C, \\
d a_i - k a_{i-1}- k a_{i+1} = 0, \\
(d - k)a_N - k x_{N-1} =0$$

ì—¬ê¸°ì—ì„œ $d=2k-mw^2$ì´ë‹¤.

ìœ„ì˜ ìš´ë™ë°©ì •ì‹ì„ í–‰ë ¬ë¡œ í‘œí˜„í•˜ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤.

$$Ma=b $$$$
M=
\begin{pmatrix}
d-k & -k & 0 & 0 & 0 & \cdots \\
-k & d & -k & 0 & 0 & \cdots \\
0 & -k & d & -k & 0 & \cdots \\
\cdots &\cdots &\cdots &\cdots &\cdots & \cdots  \\
0 & 0 & 0& 0& -k & d-k 
\end{pmatrix}, \quad
a=
\begin{pmatrix}
a_1 \\ 
a_2 \\ 
a_3 \\ 
\cdots \\ 
a_N  
\end{pmatrix}, \quad
b=
\begin{pmatrix}
C \\ 
0 \\ 
0 \\ 
\cdots \\ 
0 
\end{pmatrix}
$$

* $m=1$, $k=1$, $C=1$, $N=40$ìœ¼ë¡œ ê³ ì •í•˜ì„¸ìš”.
* $w$ (ì¦‰, d)ë¥¼ ë°”ê¾¸ì–´ê°€ë©´ì„œ $a_i$ë¥¼ êµ¬í•˜ì„¸ìš”.
* ì´ë¥¼ ê·¸ë˜í”„ë¡œ í‘œí˜„í•˜ê³  ì˜ë¯¸ë¥¼ ì„¤ëª…í•˜ì„¸ìš”. 

* 100X100 í–‰ë ¬ì„ ì§ì ‘ ì…ë ¥í•˜ëŠ” ê²ƒì„ ê¸ˆì§€í•˜ì§€ëŠ” ì•Šì§€ë§Œ ê·¸ë ‡ê²Œ í•˜ì§€ ì•Šê¸°ë¥¼ ë°”ëë‹ˆë‹¤.
* np.zeros(n) í•¨ìˆ˜ëŠ” í¬ê¸°ê°€ nì´ë©° ëª¨ë“  ì„±ë¶„ì´ 0ì¸ ë°°ì—´ì„ ë§Œë“œëŠ” ëª…ë ¹ì–´ì…ë‹ˆë‹¤.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1
k = 1
c = 1
n = 40
b = np.zeros((n,1))
b[0] = c

w = 1  # wì„ ë°”ê¾¸ì–´ ê°€ë©´ì„œ a_i êµ¬í•˜ê¸°
d = 2*k - m*w**2

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 1 ì¼ë•Œ a_i,',a)
xdata = np.arange(1,41,1)
plt.figure(figsize=(15,12))
plt.xlabel('$i$')
plt.ylabel('$a_i$')
plt.plot(xdata,a,'-',marker = 'o',label='w=1')
plt.legend()

w = 2  # wì„ ë°”ê¾¸ì–´ ê°€ë©´ì„œ a_i êµ¬í•˜ê¸°
d = 2*k - m*w**2

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 2 ì¼ë•Œ a_i,',a)
xdata = np.arange(1,41,1)
plt.plot(xdata,a,'-',marker = 'o',label='w=2')
plt.legend()

w = 3  # wì„ ë°”ê¾¸ì–´ ê°€ë©´ì„œ a_i êµ¬í•˜ê¸°
d = 2*k - m*w**3

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 3 ì¼ë•Œ a_i,',a)
xdata = np.arange(1,41,1)
plt.plot(xdata,a,'-',marker = 'o',label='w=3')
plt.legend()

w = 4  # wì„ ë°”ê¾¸ì–´ ê°€ë©´ì„œ a_i êµ¬í•˜ê¸°
d = 2*k - m*w**4

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 4 ì¼ë•Œ a_i,',a)
xdata = np.arange(1,41,1)
plt.plot(xdata,a,'-',marker = 'o',label='w=4')
plt.legend()

w = 5  # wì„ ë°”ê¾¸ì–´ ê°€ë©´ì„œ a_i êµ¬í•˜ê¸°
d = 2*k - m*w**5

for i in range(1,40):
    M[i,i] = d
    M[i,i-1] = -k
    M[i-1,i] = -k
    
M[0,0] = d-k
M[39,39] = d-k

a = np.linalg.solve(M,b)
print('w = 5 ì¼ë•Œ a_i,',a)
xdata = np.arange(1,41,1)
plt.plot(xdata,a,'-',marker = 'o',label='w=5')
plt.legend();
```
