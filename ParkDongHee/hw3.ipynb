## 1. Leibniz Pi formula


$$\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots 
\end{align}$$
이 식은 라이프니츠의 파이 공식이다.

1. 다음의 식을 구현하는 함수를 만들어라.
$$\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}$$
2. 이 식을 이용하여 아래처럼 그림을 그려라.

```python
def Leibniz_Pi_Formula(n):
    i=0
    f=0
    while(i<n):
        i=i+1
        f=f+1/((2*i-1)*(-1)**(i+1))
    return f
```

```python
n=100
x=np.arange(0,100,2)
plt.plot(x,[Leibniz_Pi_Formula(i+1) for i in range(50)], label='Leibniz formula for π')
plt.plot(x,[np.pi/4 for i in range(50)], label='π/4')
plt.xlabel("n")
plt.ylim(0.65,1)
plt.legend()
```

## 2. Logistic Map

로지스틱 맵은 다음처럼 주어진다.$$ f_{n+1} = r f_n (1- f_n). $$

1. $f_n$을 구하는 함수를 작성하여라.
2. 작성한 함수를 이용하여 다음의 조건에서 $n=1$에서 $100$까지 로지스틱 맵의 값을 구하고 그래프로 그려라. 즉, x축은 $n$, y축은 $f_n$인 그래프를 그려라.
 - $r=0.5$, $f_0$=0.5
 - $r=0.95$, $f_0$=0.5
 - $r=2$, $f_0$=0.5
 - $r=3.2$, $f_0$=0.5
 - $r=3.8$, $f_0$=0.5
 - $r=3.8$, $f_0$=0.501

```python
def Logistic_Map(r, f):
    for i in range(100):
        f=r*f*(1-f)
        print(f)
```

```python
def LogisticMap(r, f):
    x,y = [],[]
    for i in range(1,101):
        f = r*f*(1-f)
        x.append(i)
        y.append(f)


    plt.xlabel('n') 
    plt.ylabel('$f_n$', rotation=0)
    plt.plot(x, y)
    plt.show()

LogisticMap(0.5, 0.5)
LogisticMap(0.95, 0.5)
LogisticMap(0, 0.5)
LogisticMap(3.2, 0.5)
LogisticMap(3.8, 0.5)
LogisticMap(3.8, 0.501)
```

## 3. sinc function

1. 싱크 함수를 그려보아라. 싱크함수는 다음처럼 주어진다.$$ sinc(x)  = \frac{\sin x}{x}. $$
2. 싱크 함수를 테일러 전개하면 다음과 같다.$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$위의 식의 우변을 $n$을 바꾸어 가면서 그래프로 그리고 싱크함수와 비교하여라.

```python
def sinc(n):
    x=np.linspace(-10,10,n)
    y=np.sin(x)/x
    
    plt.plot(x,y)

sinc(100)
```

```python
def sinc_y(x):
    y=np.sin(x)/x
    return y
    
def sinc_Taylor(x,n):
    f = 0
    for i in range(n+1):
        f = f + ((-1)**i * x**(2*i))/(math.factorial(2*i+1))
    return f

    
x = np.arange(-10,10,0.1)
plt.plot(x, sinc_y(x), label='$sinc function$')
plt.plot(x,sinc_Taylor(x,1), label='$n=1$')
plt.plot(x,sinc_Taylor(x,4), label='$n=4$')
plt.plot(x,sinc_Taylor(x,9), label='$n=9$')
plt.plot(x,sinc_Taylor(x,16), label='$n=16$')
plt.axis([-10,10,-1,1])
plt.legend()
plt.show()
```

n이 16일때 기존의 싱크함수와 비슷한 모양을 보이고, 아래의 그래프에서 나타나지 않는걸 볼 수 있어 n이 증가할수록 형태가 싱크함수와 같아짐을 볼 수 있다.
