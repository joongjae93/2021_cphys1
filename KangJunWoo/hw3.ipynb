# 1. Leibniz Pi formula

$$\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots 
\end{align}$$
이 식은 라이프니츠의 파이 공식이다.

1. 다음의 식을 구현하는 함수를 만들어라.
$$\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}$$
2. 이 식을 이용하여 문제와 같이 그림을 그려라

```python
def f(n):
    f = 0
    for s in range(1,n+1):
        f = f + ((-1)**(s+1))*1/(2*s-1)
    return f
```

```python
a = [f(n) for n in range (1,101)]
plt.plot(a,color='black',label='Leibniz Pi',linewidth=1)
plt.hlines(np.pi/4,0,100,color='pink',label='π/4',linewidth=2)
plt.legend()
plt.grid()
```

# 2. Logistic Map
로지스틱 맵은 다음처럼 주어진다.$$ f_{n+1} = r f_n (1- f_n). $$

$f_n$을 구하는 함수를 작성하여라.
작성한 함수를 이용하여 다음의 조건에서 $n=1$에서 $100$까지 로지스틱 맵의 값을 구하고 그래프로 그려라. 즉, x축은 $n$, y축은 $f_n$인 그래프를 그려라.

$r=0.5$, $f_0$=0.5, 
$r=0.95$, $f_0$=0.5, 
$r=2$, $f_0$=0.5, 
$r=3.2$, $f_0$=0.5, 
$r=3.8$, $f_0$=0.5, 
$r=3.8$, $f_0$=0.501

```python
def LogisticMap(r,f):
    for i in range(100):
        f = r*f*(1-f)
        print('f', i+1, '=', f)
```

```python
LogisticMap(0.5,0.5)
LogisticMap(0.95,0.5)
LogisticMap(2,0.5)
LogisticMap(3.2,0.5)
LogisticMap(3.8,0.5)
LogisticMap(3.8,0.501)
```

```python
def LogisticMap(r,f):
    x, y = [], []
    for i in range(100):
        f = r*f*(1-f)
        x.append(f)
        y.append(i)
        plt.plot(y, x, linewidth=1, color='black')
```

# 3. sinc function
싱크 함수를 그려보아라. 싱크함수는 다음처럼 주어진다.$$ sinc(x)  = \frac{\sin x}{x}. $$
싱크 함수를 테일러 전개하면 다음과 같다.$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$위의 식의 우변을 $n$을 바꾸어 가면서 그래프로 그리고 싱크함수와 비교하여라.

```python
a=np.arange(-20,20,0.0001)
b=[np.sin(x)/x for x in a]
plt.plot(a,b,linewidth=2)
```

```python
def s(x,n):
    f = 0
    for i in range(n+1):
        f = f + ((-1)**i*x**(2*i))/(math.factorial(2*i+1))
    return f

x = np.arange(-20,20,0.1)
plt.plot(x, s(x,1), color='grey',linewidth=2)
plt.plot(x, s(x,4), color='blue',linewidth=2)
plt.plot(x, s(x,9), color='black',linewidth=2)
plt.plot(x, s(x,16), color='red',linewidth=2)
plt.plot(x, s(x,25), color='yellow',linewidth=2)
plt.axis([-20, 20, -2, 2])
```

n이 증가할 수록 $$ sinc(x)  = \frac{\sin x}{x}. $$ 의 그래프와 같아진다
