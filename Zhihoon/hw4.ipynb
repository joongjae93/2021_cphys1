```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate

print("#1. 오차함수")

def f(x):
   return (2/math.sqrt(math.pi))*np.exp(-x**2)

def integral(f,a,b,dx):
   s=0
   x=a
   while(x<b):
      s+=0.5*dx*(f(x+dx)+f(x))
      x=x+dx
   return s
print(integral(f,0,2,0.001))

plt.plot([i for i in range(5)],[integral(f,0,i,10**(-6)) for i in range(5)])
plt.legend('erf')
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate

def S(x,n):
    i,f = 0, 0
    for i in range(n+1):
        f=f+((-1)**i*x**(2*i+1))/((2*i+1)*math.factorial(i))
    return f

x = np.arange(0,5,0.1)
fig = plt.figure(figsize=(8,4))
plt.plot(x,S(x,10),label='Taylor if n=10')
plt.title('Taylor')
plt.xlabel('$x$')
plt.ylim([-0, 1])
plt.legend()
```

```python
print("#2. 적분")
print("I. sin 함수의 tpz.(trapezoidal) 적분")
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate

xdata = np.arange(0,8,0.01)
def yfunc(x):
    return np.sin(x)

plt.xlim(0,6);plt.ylim(-1.5,1.5)
plt.plot(xdata,yfunc(xdata));

x1 = np.arange(0,8,2)
x2 = np.arange(0,8,1)
x3 = np.arange(0,8,0.5)

plt.figure(figsize=(12,3));
plt.subplot(1,3,1)
plt.xlim(0,6);plt.ylim(-1.5,1.5)
plt.plot(x1,yfunc(x1),'o');
plt.plot(xdata,yfunc(xdata));
plt.fill_between(x1,yfunc(x1),alpha=0.2)

plt.subplot(1,3,2)
plt.xlim(0,6);plt.ylim(-1.5,1.5)
plt.plot(x2,yfunc(x2),'o');
plt.plot(xdata,yfunc(xdata));
plt.fill_between(x2,yfunc(x2),alpha=0.2)

plt.subplot(1,3,3)
plt.xlim(0,6);plt.ylim(-1.5,1.5)
plt.plot(x3,yfunc(x3),'o');
plt.plot(xdata,yfunc(xdata));
plt.fill_between(x3,yfunc(x3),alpha=0.2)

def yfunc(x):
    return np.cos(x)

def trapez(f,a,b,n):
    s, x, dx = 0, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*f(x)+0.5*f(x+dx))
        x = x + dx
    return s

trapez(yfunc,0,6,100)
```

```python
print("#3. 미분")
print("I. sin 함수의 tpz. 미분")

import matplotlib.pyplot as plt
import numpy as np

xdata = np.arange(0,8,0.01)
def yfunc(x):
    return np.sin(x)

plt.xlim(0,6);plt.ylim(-1.5,1.5)
plt.plot(xdata,yfunc(xdata));

def trapez(f,a,b,n):
    s, x, dx = 0, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*f(x)+0.5*f(x+dx))
        x = x + dx
    return s

def for_diff(f,x,dx):
    return (f(x+dx)-f(x))/dx
def back_diff(f,x,dx):
    return (f(x)-f(x-dx))/dx
def cen_diff(f,x,dx):
    return (f(x+dx/2)-f(x-dx/2))/dx

dx=0.1
xdata=np.arange(0,8,0.01)
diff1=[for_diff(yfunc,i,dx) for i in xdata]
diff2=[back_diff(yfunc,i,dx) for i in xdata]
diff3=[cen_diff(yfunc,i,dx) for i in xdata]

plt.plot(xdata,diff1)
plt.plot(xdata,diff2)
plt.plot(xdata,diff3)
plt.plot(xdata,yfunc(xdata));
```

```python
print("II. cos 함수의 tpz. 미분")

import matplotlib.pyplot as plt
import numpy as np

xdata = np.arange(0,8,0.01)
def yfunc(x):
    return np.cos(x)

plt.xlim(0,6);plt.ylim(-1.5,1.5)
plt.plot(xdata,yfunc(xdata));

def trapez(f,a,b,n):
    s, x, dx = 0, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*f(x)+0.5*f(x+dx))
        x = x + dx
    return s

def for_diff(f,x,dx):
    return (f(x+dx)-f(x))/dx
def back_diff(f,x,dx):
    return (f(x)-f(x-dx))/dx
def cen_diff(f,x,dx):
    return (f(x+dx/2)-f(x-dx/2))/dx

dx=0.1
xdata=np.arange(0,8,0.01)
diff1=[for_diff(yfunc,i,dx) for i in xdata]
diff2=[back_diff(yfunc,i,dx) for i in xdata]
diff3=[cen_diff(yfunc,i,dx) for i in xdata]

plt.plot(xdata,diff1)
plt.plot(xdata,diff2)
plt.plot(xdata,diff3)
plt.plot(xdata,yfunc(xdata));
```

```python
print("III. sin 함수의 S.M.D 미분")
from scipy.misc import derivative
derivative(np.sin,np.pi,dx=0.01)
plt.plot(xdata,[derivative(np.sin,i,dx=0.01) for i in xdata])
```
