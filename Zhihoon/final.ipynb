```python
import numpy as np
import matplotlib.pyplot as plt
 
N = 115 # number of random walkers
T = 15  # time span
 
position = np.zeros((N, T+1))
 
for t in range(1, T+1):
    for i in range(N):
        position[i][t] = position[i][t-1] + np.random.normal()
 
plt.figure()
for i in range(N):
    plt.plot(position[i])
 
mean = np.mean(position, axis=0)
std = np.std(position, axis=0)
 
plt.figure()
plt.plot(mean,"red")
plt.plot(std,"blue")
```

```python
tdata = np.arange(0,8,0.01)

def x1(t):
    return -2/(2-3*math.e**t)
def x2(t):
    return -2/(2-math.e**t)
def x3(t):
    return 1/(1-math.e**(-t)/2)
def x4(t):
    return 1/(1-3*math.e**(-t)/2)
```

```python
plt.figure(figsize=(15,3));

plt.subplot(1,4,1)
plt.xlim(0,6);plt.ylim(0,2.5)
plt.plot(tdata,x1(tdata));
plt.legend("1")

plt.subplot(1,4,2)
plt.xlim(0,6);plt.ylim(0,2.5)
plt.plot(tdata,x2(tdata));
plt.legend("2")

plt.subplot(1,4,3)
plt.xlim(0,6);plt.ylim(0,2.5)
plt.plot(tdata,x3(tdata));
plt.legend("3")

plt.subplot(1,4,4)
plt.xlim(0,6);plt.ylim(0,2.5)
plt.plot(tdata,x4(tdata));
plt.legend("4")
```
