```python
print("#1.")
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

w=1
g=9.8
L=9.8
state0 = [1.0, 0.0]
t = np.arange(0.0, 10.0, 0.01)

def f(state, t):
    theta, w = state 
    return w, -(g/L)*math.sin(theta)

states = odeint(f, state0, t)
plt.plot(t,states[:,0])
```
