## 1. Vibration in a one-dimensional system

질량이 $m$인 $N$개의 동일한 물체가 일렬로 동일한 용수철에 매달려 있다. $i$번째 물체의 운동방정식은 다음처럼 주어진다.
$$m \frac{d^2 x_i}{dt^2} = k(x_{i+1}-x_{i}) - k(x_i-x_{i-1})$$
한쪽 끝에서 연결된 용수철에 주기적인 힘($F$)을 가하고 있다. 그러면 양 끝단에 물체에 운동방정식은 다음과 같다.
$$m \frac{d^2 x_1}{dt^2} = k(x_{2}-x_{1}) + F \\
m \frac{d^2 x_N}{dt^2} = -k(x_{N}-x_{N-1})$$
만약 주기적인 힘이 $F=C e^{i wt}$로 주어지면 $x$는 다음의 $x_i(t)=a_i e^{iwt}$해를 가진다. 정리하면 다음과 같은 방정식을 얻을 수 있다.
$$(d -k)a_1 -k a_2 = C, \\
d a_i - k a_{i-1}- k a_{i+1} = 0, \\
(d - k)a_N - k x_{N-1} =0$$
여기에서 $d=2k-mw^2$이다.

위의 운동방정식을 행렬로 표현하면 다음과 같다.
$$Ma=b $$$$
M=
\begin{pmatrix}
d-k &amp; -k &amp; 0 &amp; 0 &amp; 0 &amp; \cdots \\
-k &amp; d &amp; -k &amp; 0 &amp; 0 &amp; \cdots \\
0 &amp; -k &amp; d &amp; -k &amp; 0 &amp; \cdots \\
\cdots &amp;\cdots &amp;\cdots &amp;\cdots &amp;\cdots &amp; \cdots  \\
0 &amp; 0 &amp; 0&amp; 0&amp; -k &amp; d-k 
\end{pmatrix}, \quad
a=
\begin{pmatrix}
a_1 \\ 
a_2 \\ 
a_3 \\ 
\cdots \\ 
a_N  
\end{pmatrix}, \quad
b=
\begin{pmatrix}
C \\ 
0 \\ 
0 \\ 
\cdots \\ 
0 
\end{pmatrix}
$$

$m=1$, $k=1$, $C=1$, $N=40$으로 고정하세요.
$w$ (즉, d)를 바꾸어가면서 $a_i$를 구하세요.
이를 그래프로 표현하고 의미를 설명하세요. x축이 i (1~N), y축이 $a_i$

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg
```

```python
#w=1
m,k,w=1,1,1

def d(w):
    return (2*k)-(m*(w**2))

M1 = np.identity(40)
for N in range(0,40):
    for M in range(0,40):
        if N == 0 and M == 0:
            M1[N][M] = d(w)-k
            
        elif N == 0 and M == 1:
            M1[N][M] =-k
            
        elif N == 39 and M == 38:
            M1[N][M] =-k
            
        elif N == 39 and M == 39:
            M1[N][M] = d(w)-k
            
        elif N == M+1:
            M1[N][M] =-k
            
        elif N == M:
            M1[N][M] = d(w)
            
        elif N == M-1:
            M1[N][M] =-k
print(M1)
M1_i = np.linalg.inv(M1)
print(M1_i)
```

```python
#w=2
m,k,w=1,1,2

def d(w):
    return (2*k)-(m*(w**2))

M2 = np.identity(40)
for N in range(0,40):
    for M in range(0,40):
        if N == 0 and M == 0:
            M2[N][M] = d(w)-k
            
        elif N == 0 and M == 1:
            M2[N][M] =-k
            
        elif N == 39 and M == 38:
            M2[N][M] =-k
            
        elif N == 39 and M == 39:
            M2[N][M] = d(w)-k
            
        elif N == M+1:
            M2[N][M] =-k
            
        elif N == M:
            M2[N][M] = d(w)
            
        elif N == M-1:
            M2[N][M] =-k
print(M2)
M2_i = np.linalg.inv(M2)
print(M2_i)
```

```python
#w=3
m,k,w=1,1,3

def d(w):
    return (2*k)-(m*(w**2))

M3 = np.identity(40)
for N in range(0,40):
    for M in range(0,40):
        if N == 0 and M == 0:
            M3[N][M] = d(w)-k
            
        elif N == 0 and M == 1:
            M3[N][M] =-k
            
        elif N == 39 and M == 38:
            M3[N][M] =-k
            
        elif N == 39 and M == 39:
            M3[N][M] = d(w)-k
            
        elif N == M+1:
            M3[N][M] =-k
            
        elif N == M:
            M3[N][M] = d(w)
            
        elif N == M-1:
            M3[N][M] =-k
print(M3)
M3_i = np.linalg.inv(M3)
print(M3_i)
```

```python
C=np.eye(40,1,0)

M_1= np.dot(M1_i,C)
M_2 = np.dot(M2_i,C)
M_3 = np.dot(M3_i,C)

plt.plot(M_1,label='w=1')
plt.plot(M_2,label='w=2')
plt.plot(M_3,label='w=3')
plt.xlabel('$a_i$')
plt.ylabel('$i$')
```

w가 증가하면 x 또한 증가하게 되어 결론적으로 ai값이 0에 수렴하게 된다.

## 2. 양쪽 끝에서 힘을 가하면 어떻게 되는가?

즉, b행렬이 다음과 같은 경우에 대해서 살펴보아라.$$
b=
\begin{pmatrix}
C \\ 
0 \\ 
0 \\ 
\cdots \\ 
C 
\end{pmatrix}
$$

$m=1$, $k=1$, $C=1$, $N=40$으로 고정하세요.
$w$ (즉, d)를 바꾸어가면서 $a_i$를 구하세요.
이를 그래프로 표현하고 의미를 설명하세요.

```python
C = np.eye(40,1,0)
CP = np.flipud(C)
C_1 = CP + C
```

```python
M_1 = np.dot(M1_i,C_1)
M_2 = np.dot(M2_i,C_1)
M_3 = np.dot(M3_i,C_1)

plt.plot(M_1,label='w=1')
plt.plot(M_2,label='w=2')
plt.plot(M_3,label='w=3')
plt.xlabel('$a_i$')
plt.ylabel('$i$')
```

w가 증가하면 양쪽에서 힘을 받게 되기 때문에 결론적으로 중앙에서 0으로 수렴하게 된다.

w가 증가하면 x 또한 증가하게 되어 결론적으로 ai값이 0에 수렴하게 된다. <- x 가 무엇인가요?

w = 1 인 경우도 양쪽에서 힘을 주는 건 똑같은데 왜 중앙에서 수렴하지 않나요? (-1.5)
