1.Pendulum

```python
import numpy as np
g,L=9.8,9.8
a = g/L
dt=0.001
t = np.arange(0,100,dt)
theta = np.array([])
w = np.array([])
theta0, w0 = 1, 0


for i in t:
    theta0 = theta0 + dt*w0
    w0 = w0 - a*np.sin(theta0)*dt
    theta=np.append(theta,theta0)
    w=np.append(w,w0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(t,theta)
```

3.Damped Pendulum

```python
import numpy as np
beta, g, L=0.1, 9.8, 9.8
a = g/L 
dt=0.01
t = np.arange(0,100,dt)
theta = np.array([])
w = np.array([])
theta0, w0 = 1, 0


for i in t:
    theta0 = theta0 + dt*w0
    w0 = w0 - beta*w0*dt-a*np.sin(theta0)*dt
    theta=np.append(theta,theta0)
    w=np.append(w,w0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(t,theta)
```

4. Forced Pendulum

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

beta, v, L, g, A=0.1, 1, 9.8, 9.8, 1
a=g/L
state0 = [1.0, 0.0]
t = np.arange(0.0, 100.0, 0.01)

def f(state, t):
    theta, w = state 
    return w, -beta*w-a*np.sin(theta)+A*np.cos(v*t)

states = odeint(f, state0, t)
plt.plot(t,states[:,0])
```
