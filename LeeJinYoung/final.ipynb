Í∑∏Î¶¨Í≥† $t$ÏãúÍ∞ÑÏù¥ ÏßÄÎÇú ÌõÑÏóê ÏûÖÏûêÍ∞Ä Ï°¥Ïû¨ÌïòÎäî ÏúÑÏπòÎ•º Íµ¨ÌïòÎäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏó¨Îùº. Ï¥àÍ∏∞Ïóê ÏûÖÏûêÎäî ÏõêÏ†êÏóê Ï°¥Ïû¨ÌïòÎ©∞ Îß§ ÌÉÄÏûÑÏä§ÌÖùÎßàÎã§ 1Î≤àÏî© ÏõÄÏßÅÏù∏Îã§.
```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sympy import *
import sympy
```

```python
def Randomwalks(n):
    x = 0
    y = 0
    time = [x]
    position = [y]
    
    for i in range (1,n+1):
        step = np.random.uniform(0,1)
        if step < 0.5:
            x += 1
            y += 1
            
        if step > 0.5:
            x += 1
            y += -1
        
        time.append(x)
        position.append(y)
    return [time, position]

Randomwalks = Randomwalks(1000)
plt.plot(Randomwalks[0],Randomwalks[1],label="Randomwalks")
```

(b) Ïù¥Îü¨Ìïú ÏãúÌñâÏùÑ Ï∂©Î∂ÑÌûà ÎßéÏù¥ ÏãúÌñâÌïòÏó¨ÏÑú  ùë° ÏãúÍ∞Ñ ÌõÑÏóê ÏûÖÏûêÍ∞Ä ÌäπÏ†ï ÏúÑÏπòÏóê Ï°¥Ïû¨ÌïòÎäî ÌôïÎ•† Î∂ÑÌè¨Î•º Íµ¨ÌïòÏó¨Îùº. (ÌååÏù¥Ïç¨ Í∏∞Ï¥à)

```python
def Randomwalks(n):
    x = 0
    y = 0
    time = [x]
    position = [y]
    
    for i in range (1,n+1):
        step = np.random.uniform(0,1)
        if step < 0.5:
            x += 1
            y += 1
            
        if step > 0.5:
            x += 1
            y += -1
        
        time.append(x)
        position.append(y)
    return [time,position], position[-1]

N = 10000
n = 100
X = []
for i in range(N):
    X.append(Randomwalks(n)[1])

plt.hist(X, bins=80, density = True, alpha = 1)
plt.show
```

(a) Ïù¥ Î∞©Ï†ïÏãùÏùÑ Îã§ÏùåÏùò Ï°∞Í±¥ÏóêÏÑú ÌíÄÍ≥† Í∂§Ï†ÅÏùÑ Í∑∏Î†§Îùº. (ÎØ∏Î∂ÑÎ∞©Ï†ïÏãù)

```python
x = symbols('x', cls=Function)
r = symbols('r')
t = symbols('t')
x(t)
```

```python
Eq(x(t).diff(t), r * x(t) - x(t)**2)
```

```python
r= -1
deq1 = Eq(x(t).diff(t), r* x(t) - x(t)**2)
psol1 = dsolve(deq1, ics={x(0):2})
psol1
```

```python
plot(psol1.rhs,(t,0,10))
```

```python
r= -1
deq2 = Eq(x(t).diff(t), r* x(t) - x(t)**2)
psol2 = dsolve(deq2, ics={x(0):-2})
psol2
```

```python
plot(psol2.rhs,(t,0.7,10))
```

```python
r= 1
deq3 = Eq(x(t).diff(t), r* x(t) - x(t)**2)
psol3 = dsolve(deq3, ics={x(0):2})
psol3
```

```python
plot(psol3.rhs,(t,0,10))
```

```python
r= 1
deq4 = Eq(x(t).diff(t), r* x(t) - x(t)**2)
psol4 = dsolve(deq4, ics={x(0):-2})
psol4
```
